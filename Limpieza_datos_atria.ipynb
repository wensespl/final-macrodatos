```python
import pandas as pd
```

```python
df = pd.read_csv('/content/datos_atria.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Especificar las columnas que deseas eliminar
columnas_a_eliminar = ['rubro', 'ventas']

# Eliminar las columnas especificadas
df = df.drop(columns=columnas_a_eliminar)
```

```python
# Renombrar una columna
df = df.rename(columns={'energia': 'costo_energia'})
```

```python
df.head()
```

```python
# Definimos la función para calcular la diferencia en días entre dos fechas
def calcular_diferencia_dias(row):
    inicio = pd.to_datetime(row['inicio_suministro'], format='%d/%m/%Y')
    fin = pd.to_datetime(row['fin_suministro'], format='%d/%m/%Y')
    return (fin - inicio).days
```

```python
# Aplicar la función usando lambda y apply para rellenar plazo_contrato
df['plazo_contrato'] = df.apply(lambda row: calcular_diferencia_dias(row), axis=1)
```

```python
# Calcular el promedio de los valores positivos en la columna precio_potencia
promedio_precio_potencia = df[df['precio_potencia'] > 0]['precio_potencia'].mean()

# Definir función lambda para cambiar valores -1 o nulos al promedio de los positivos en precio_potencia
df['precio_potencia'] = df['precio_potencia'].apply(lambda x: promedio_precio_potencia if pd.isnull(x) or x == -1 else x)
```

```python
# Definimos una función para aplicar cambio de moneda a una columna específica
def aplicar_cambio_moneda(row, columna):
    if row['moneda'] == 'PEN' and pd.notna(row[columna]):
        return row[columna] / 3.8474
    else:
        return row[columna]

# Aplicar la función usando lambda y apply para modificar precio_potencia
df['precio_potencia'] = df.apply(lambda row: aplicar_cambio_moneda(row, 'precio_potencia'), axis=1)
df['precio_energia'] = df.apply(lambda row: aplicar_cambio_moneda(row, 'precio_energia'), axis=1)
```

```python
# Definir función para realizar los cálculos según la unidad de precio de energía
def calcular_precio_energia(row):
    if row['unidad_precio_energia'] == 'kWh':
        return row['precio_energia'] * 1000
    elif row['unidad_precio_energia'] == 'kWm':
        return row['precio_energia'] * 1000 / 720
    else:
        return row['precio_energia']

# Aplicar la función usando lambda y apply para modificar precio_energia
df['precio_energia'] = df.apply(lambda row: calcular_precio_energia(row), axis=1)
```

```python
# Eliminar la columna 'moneda' del DataFrame
df = df.drop(columns=['moneda'])
```

```python
from datetime import datetime
# Obtener la fecha actual
fecha_actual = datetime.now()

# Función booleana para verificar si la fecha actual es menor, igual o mayor que fin_suministro
def comparar_fecha_actual_vs_fin_suministro(row):
    fecha_fin = datetime.strptime(row['fin_suministro'], '%d/%m/%Y')
    return fecha_actual <= fecha_fin

# Aplicar la función usando lambda y apply para obtener un booleano
df['vigencia_hoy'] = df.apply(lambda row: comparar_fecha_actual_vs_fin_suministro(row), axis=1)
```
